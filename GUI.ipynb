```python
from ipywidgets import interact,interactive,fixed
import ipywidgets as widgets
```

```python
def func(x):
    return x
```

```python
interact(func,x=10)
```

```python
interact(func,x=True)
```

```python
interact(func,x='Hello')
```

```python
@interact(x=True, y=1.0)
def g(x,y):
    return(x,y)
```

```python
@interact(x=True, y=fixed(1.0))
def g(x,y):
    return(x,y)
```

```python
interact(func,x=widgets.IntSlider(min=100,max=100, step=1, valuue=0))
```

```python
interact(func,x=(-100,100,1))
```

```python
interact(func,x=(-10.0,10.0,.1))
```

```python
@interact(x=(0.0,20.0,0.5))
def h(x=5.0):
    return x
```

```python
interact(func,x=['hello','option 2','option 3'])
```

```python
interact(func,x={'one':10, 'two':20})
```

### WIDGET TYLING AND LAYOUT

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
w = widgets.IntSlider()
display(w)
```

```python
w.layout.margin = 'auto'
w.layout.height = '75px'
```

```python
x = widgets.IntSlider(value=15, description = 'New Slider')
display(x)
```

```python
x.layout = w.layout
```

```python
widgets.Button(description = 'Ordinary Button', button_style='info')
```

```python
b1 = widgets.Button(description = 'Custom Color')
b1.style.button_color = 'lightgreen'
b1
```

```python
b1.style.keys
```

```python
b2 = widgets.Button(description = 'New')
b2.style = b1.style
b2
```

```python
s1 = widgets.IntSlider(description = 'My handle')
s1.style.handle_color = 'blue'
s1
```

```python
s1.style.keys
```
