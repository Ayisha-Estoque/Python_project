```python
try:
    f = open ('testfile','r')
    f.write('Write a test line')

except OSError:
    print('All other exceptions!')
finally:
    print('I always run')
```

```python
def ask_for_int():
    while True:

        try:
            result = int (input('Please provide a number: '))
        except:
            print('Whoops! that is not a number')
            continue
        else: 
            print('Yes thank you')
            break
        finally:
            print('End of try/except/finally')
            print('I will always run at the end!')
```
