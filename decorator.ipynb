```python
def hello(name = 'Jose'):
    print('The hello() function has been executed')

    def greet():
        return '\t This is the greet() func inside hello!'
    
    def welcome():
        return '\t This is welcome() inside hello'
    
    print("I am going  to return a function!")

    if name == 'Jose':
        return greet
    else:
        return welcome
```

```python
my_new_func = hello ('Jose')
```

```python
print(my_new_func())
```

```python
hello()
```

```python
welcome()
```

```python
def new_decorator(original_func):

    def wrap_func():
        print('Some extra code, beofre the original function')

        original_func()

        print("Some extra code, sfter the original Function")

    return wrap_func
```

```python
def func_needs_decorator():
    print("I want to be decorator!!")

func_needs_decorator()
```

```python
decorated_func = new_decorator(func_needs_decorator)
```

```python
decorated_func()
```

```python
@new_decorator
def func_needs_decorator():
    print("I want to be decorator!!")

func_needs_decorator()
```
