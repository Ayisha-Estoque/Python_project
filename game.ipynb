SHUFFLE FUNCTION

```python
example = [1,2,3,4,5,6,7]
```

```python
from random import shuffle
```

```python
shuffle(example)
```

```python
def shuffle_list (mylist):
    shuffle(mylist)
    return mylist
```

```python
result = shuffle_list(example)

result
```

```python
mylist = [' ', 'O', '']
```

```python
def player_guess():

    guess = ''

    while guess not in ['0', '1', '2']:
        guess = input("Pick a number: 0, 1, or 2")

    return int(guess)
```

```python
player_guess()
```

```python
def check_guess(mylist,guess):
    if mylist[guess] =='0':
        print("Correct!")
    else:
        print("Wrong guess!")
        print(mylist)
```

```python
mylist = [' ', '0', ' ']

mixedup_list = shuffle_list(mylist)

guess = player_guess()

check_guess(mixedup_list, guess)
```
