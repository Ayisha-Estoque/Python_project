BASIC CALCULATOR

```python
def sum_numbers(num1, num2, operator):
    if operator == '+':
        return num1 + num2
    elif operator == '-':
        return num1 - num2
    elif operator == '*':
        return num1 * num2
    elif operator =='/':
        return num1 / num2
    else:
        print('Error operator!!')

num1 = int(input ("Enter the first number: "))
num2 = int(input ("Enter the second number: "))
operator = input('Enter your operator: ')

result = sum_numbers(num1, num2, operator)

result
```
