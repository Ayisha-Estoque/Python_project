```python
file_path = 'c:\\Users\\ayish\\Downloads\\Python_project\\Complete-Python-3-Bootcamp\\12-Advanced Python Modules\\Example_Top_Level'
```

```python
for folder, sub_folders, files in os.walk(file_path):
    print(f"Currently looking at (folder)")
    print('\n')
    print('The subfolders are: ')
    for sub_fold in sub_folders:
        print(f"\t Subfolder: {sub_fold}")
    print('\n')
    print("The files are: ")
    for f in files:
        print(f'File: {f}')
    print('\n')
```
