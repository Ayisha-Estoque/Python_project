```python
def create_cubes(n):

    for x in range(n):
        yield x**3
```

```python
for x in create_cubes(10):
    print (x)
```

```python
list(create_cubes(10))
```

```python
def gen_fibon(n):

    a = 1
    b = 1
    output = []

    for i in range(n):
        output.append(a)
        a,b = b, a+b
    return output
```

```python
for number in gen_fibon(10):
    print(number)
```

```python
def simple_gen():
    for x in range(3):
        yield x
```

```python
for number in simple_gen():
    print(number)
```
