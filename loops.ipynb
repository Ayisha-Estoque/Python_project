```python
mystring = 'Sammy'
```

```python
for letter in mystring:
    if letter == 'a':
        continue
        #goes to the top of the closest enclosing loop
    print(letter)
```

```python
for letter in mystring:
    if letter == 'a':
        break
        #breaks out of the current closest enclosing loop
    print(letter)
```

```python
x = 0

while x < 5:

    if x == 2:
        break
    print(x)
    x += 1
```
