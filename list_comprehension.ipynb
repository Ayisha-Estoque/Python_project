MAKING A LIST

```python
mystring = 'hello'
```

```python
mylist = []

for letter in mystring:
    mylist.append(letter)
```

```python
celcius = [0, 10, 20, 34.5]

fahrenheit = [((9/5)*temp + 32) for temp in celcius]
```

```python
fahrenheit
```

```python
fahrenheit = []

for temp in celcius:
    fahrenheit.append(((9/5)*temp + 32))
```

```python
fahrenheit
```
