```python
class Circle():
    pi = 3.14

    def __init__(self, radius = 1):

        self.radius = radius
        self.area = radius * radius * self.pi

    def get_circumference(self):
        return self.radius * self.pi * 2
```

```python
my_circle = Circle(30)
```

```python
my_circle.pi
```

```python
my_circle.radius   
```

```python
my_circle.get_circumference()
```

INHERITANCE -new classes that is already defined

```python
class Animal():
    def __init__(self):
        print('Animal Created')
    
    def who_am_i(self):
        print('I am an animal')
    
    def eat(self):
        print('I am eating')
```

```python
class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print('Dog Created')
    
    def who_am_i(self):
        print("I am a dog!")
    
    def bark (self):
        print("WOOF!")
```

```python
mydog = Dog()
```

```python
mydog.eat()
```

```python
mydog.who_am_i()
```

```python
mydog.bark()
```

```python
myanimal = Animal()
```

```python
myanimal.eat()
```

```python
myanimal.who_am_i()
```

POLYMORPHISM

```python
class Dog():

    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return self.name + " says woof!"
```

```python
class Cat():

    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return self.name + " says meow!"
```

```python
niko = Dog ("niko")
felix = Cat ("felix")
```

```python
print(niko.speak())
```

```python
print(felix.speak())
```

```python
for pet in [niko, felix]:

    print(type(pet))
    print(type(pet.speak()))
```

```python
def pet_speak(pet):
    print(pet.speak())
```

```python
pet_speak(niko)
```

```python
pet_speak(felix)
```

```python
class Animal():
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError ("Subclass must implemented this abstract method")
```

```python
class Dog (Animal):
    def speak (self):
        return self.name + " says woof!"
```

```python
class Cat (Animal):
    def speak (self):
        return self.name + " says meow!"
```

```python
fido = Dog ("Fido")
isis = Cat ("Isis")
```

```python
print (fido.speak())
```

```python
print (isis.speak())
```

```python
class Book():

    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        return f"{self.title} by {self.author}"
    
    def __len__(self):
        return self.pages
    
    def __del__(self):
        print ("A book object has been deleted")
```

```python
b = Book ('Python rocks', 'Jose', 200)
```

```python
print(b)
```

```python
str(b)
```

```python
len(b)
```

```python
del b
```
